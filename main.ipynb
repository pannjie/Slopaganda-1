```python
import itertools 
from collections import defaultdict
```

```python
cols = mega_df_simple.columns.tolist()
cols.insert(0, cols.pop(cols.index("target_username")))
mega_df_simple = mega_df_simple[cols]
mega_df_simple.head()
```

```python
target_followers = defaultdict(set)

for _, row in mega_df_simple.iterrows():
    target_followers[row['target_username']].add(row['screen_name'])

target_followers
```

```python
results = []
for t1, t2 in itertools.combinations(target_followers, 2):
   shared = len(target_followers[t1].intersection(target_followers[t2]))
   if shared >= 3:
     results.append((t1, t2, shared))

results.sort(key=lambda x: x[2], reverse=True)
results[:10]
```

```python
len(results)
df_results = pd.DataFrame(results, columns=["target1", "target2", "shared_followers"])
df_results
```

```python
df_results.to_csv("shared_followers.csv", index=False)
```

```python
df_final = df_results.merge(mega_df_simple[["target_username", "followers_count","created_at"]].drop_duplicates(), left_on="target1", right_on="target_username").rename(columns={"followers_count": "target1_followers"}).drop(columns=["target_username"])

df_final = df_final.drop_duplicates(subset=["target1", "target2"])

df_final
```

```python
g = rp.Graph()

g.load_nodes(
  data=df_final,
  id="target1",
  time="created_at",
)

g.load_edges(
  data=df_final,
  src="target2",
  dst="target1",
  time="created_at",
)
print(g)
```

```python
# Edge density of df_final
unique_nodes = pd.concat([df_final['target1'], df_final['target2']]).nunique()
unique_edges = df_final[['target1', 'target2']].drop_duplicates().shape[0]

edges_per_node = unique_edges / unique_nodes
# Directed graph density: edges / (n * (n - 1))
directed_density = unique_edges / (unique_nodes * (unique_nodes - 1))

print(f"Unique nodes:       {unique_nodes}")
print(f"Unique edges:       {unique_edges}")
print(f"Edges per node:     {edges_per_node:.4f}")
print(f"Directed density:   {directed_density:.6f}")
```

```python
clustering_coefficient = rp.algorithms.global_clustering_coefficient(g)
print(f"The graph's clustering coefficient is {clustering_coefficient}")
```

```python
clustering = algorithms.louvain(g)
```

```python
# Inspect structure first
sample_node, sample_val = next(iter(clustering.items()))
print(f"Cluster value structure: {sample_val}")

# Extract the actual cluster ID from the dict
unique_clusters = {next(iter(cluster.values())) for node, cluster in clustering.items()}
print(f"\nNumber of unique clusters: {len(unique_clusters)}")
```

```python
labels = {0: 'Team_1', 1: 'Team_2', 2: 'Team_3'}
for node, cluster in clustering.items():
    cluster_id = next(iter(cluster.values()))          # fix 1: extract ID from dict
    g.node(node).set_node_type(labels[cluster_id])

print("Assigned node types based on clusters:")
for label in labels.values():
    nodes_in_cluster = [n.name for n in g.nodes if n.node_type == label]  # fix 2: g not G
    print(f"  {label}: {len(nodes_in_cluster)} nodes")

# Start a GraphQL server to explore in the UI
work_dir = tempfile.mkdtemp()
server = GraphServer(work_dir).start()                 # fix 3: GraphServer not graphql.GraphServer
client = server.get_client()
client.send_graph("cluster-graph", g, overwrite=True)
print(f"\nServer running at http://localhost:1736")
print(f"Open the Raphtory UI to explore the clustered graph")
```

```python
from collections import defaultdict

# Group nodes by their cluster ID
communities = defaultdict(list)
for node, cluster in clustering.items():
    cluster_id = next(iter(cluster.values()))  # or cluster["community"] etc.
    communities[cluster_id].append(node)

# Print each community and its members
for cluster_id, members in sorted(communities.items()):
    print(f"\nCluster {cluster_id} ({len(members)} nodes):")
    for m in members:
        print(f"  {m}")
```
